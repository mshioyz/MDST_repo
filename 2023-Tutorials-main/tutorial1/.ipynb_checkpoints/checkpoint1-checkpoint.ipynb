<h2> Data Cleanup </h2>

Import `numpy`, `pandas`, and `matplotlib`.

(Feel free to import other libraries!)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
df = pd.read_csv('../data/states_edu.csv')
df.head()
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
df.rename({
    'GRADES_PK_G':'ENROLL_PREK',
    'GRADES_KG_G':'ENROLL_KINDER',
    'GRADES_4_G':'ENROLL_4',
    'GRADES_8_G':'ENROLL_8',
    'GRADES_12_G':'ENROLL_12',
    'GRADES_1_8_G':'ENROLL_PRIMARY',
    'GRADES_9_12_G':'ENROLL_HS',
    'GRADES_ALL_G':'ENROLL_ALL',
    'ENROLL':'ENROLL_ALL_EST'
    },
    axis=1,inplace=True)
df['ENROLL_ALL'] = df['ENROLL_ALL'].fillna(df["ENROLL_PREK"]+df["ENROLL_PRIMARY"]+df["ENROLL_HS"])
df["ENROLL_ALL_EST"] = df["ENROLL_ALL_EST"].fillna(df["ENROLL_ALL"])
df.dropna(subset=['AVG_MATH_8_SCORE'], inplace=True)
```

<h2>Exploratory Data Analysis (EDA) </h2>

Chosen Outcome Variable for Test: Grade 8 Math

How many years of data are logged in our dataset? 

```python
years = df["YEAR"].nunique()
print(years)
```

Let's compare Michigan to Ohio. Which state has the higher average outcome score across all years?

```python
state_score = df.groupby("STATE")["AVG_MATH_8_SCORE"].mean()
michigan_ave_score = state_score.loc["MICHIGAN"]
ohio_ave_score = state_score.loc["OHIO"]

if michigan_ave_score > ohio_ave_score:
    print("Michigan has a higher average math score for 8th graders.")
elif ohio_ave_score > michigan_ave_score:
    print("Ohio has a higher average math score for 8th graders.")
else:
    print("Michigan and Ohio have the same average math score for 8th graders.")
```

Find the average for your outcome score across all states in 2019

```python
average_2019 = df[df["YEAR"] == 2019]["AVG_MATH_8_SCORE"].mean()
print(average_2019)
```

Find the maximum outcome score for every state. 

Refer to the `Grouping and Aggregating` section in Tutorial 0 if you are stuck.

```python
max_scores_by_state = df.groupby("STATE")["AVG_MATH_8_SCORE"].max()
print(max_scores_by_state)
```

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df["EXPENDITURE_PER_ENROLL"] = df["TOTAL_EXPENDITURE"]/df["ENROLL_ALL"]
df["FEDERAL_REVENUE_PROPORTION"] = df["FEDERAL_REVENUE"] / df["TOTAL_REVENUE"]
df["STATE_REVENUE_PROPORTION"] = df["STATE_REVENUE"] / df["TOTAL_REVENUE"]
df["LOCAL_REVENUE_PROPORTION"] = df["LOCAL_REVENUE"] / df["TOTAL_REVENUE"]
df["INSTRUCTION_EXPENDITURE_PROPORTION"] = df["INSTRUCTION_EXPENDITURE"] / df["TOTAL_EXPENDITURE"]

df.sort_values(by='YEAR', inplace=True)
df['PERCENT_CHANGE_MATH_8_SCORE'] = df['AVG_MATH_8_SCORE'].pct_change() * 100
df.reset_index(drop=True, inplace=True)

# teacher_per_student_ratio = teacher_count / "ENROLL_ALL"
# region eg West, Midwest, South, Northeast
# household_income
```

Feature engineering justification: **<BRIEFLY DESCRIBE WHY YOU MADE THE CHANGES THAT YOU DID\>**

EXPENDITURE_PER_ENROLLn: Higher expenditure per studend indicatse better-funded schools, potentially leading to improved resources, smaller class sizes, and better educational outcomes

Proportion of revenue: Depending on the sources of funding, the government and states have different levels of control on schools regulations or curriculum.

INSTRUCTION_EXPENDITURE_PROPORTION: This may show the level of focus that the school have on teachers, which may indicate better paid and therefore teachers of better quality.

PERCENT_CHANGE_MATH_8_SCORE: Calculate the percentage change in score on a yoy basis. This allows for observation of the trend in academic performance over time for that school, which might have a correlation with the grades of the current year.

teacher_per_student_ratio: indicates class size and amount of attention each student receives, allowing for more personalised feedback and potentially better educational outcomes

region: different regions may have different educational policies which affects the curriculum.

household_income: this affects the level of resources each student gets both inside and outside the classroom, as well as the extent which the studnet can prioritize academic affairs over daily affairs eg putting food on the table, which affects educational outcomes. .

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
cols = ['EXPENDITURE_PER_ENROLL','INSTRUCTION_EXPENDITURE_PROPORTION', 'FEDERAL_REVENUE_PROPORTION', 'PERCENT_CHANGE_MATH_8_SCORE']
X = df[cols].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
      X, y, test_size=.2, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.intercept_)
print(model.coef_)
print(model.score(X_test, y_test))
print(np.mean(model.predict(X_test)-y_test))
print(np.mean((model.predict(X_test)-y_test)**2)**0.5)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
for col_name in cols:
    f = plt.figure(figsize=(12,6))
    plt.scatter(X_train[col_name], y_train, color = "red")
    plt.scatter(X_train[col_name], model.predict(X_train), color = "green")
    
    plt.legend(['True Training','Predicted Training'])
    plt.xlabel(col_name)
    plt.ylabel('AVG_MATH_8_SCORE')
    plt.title("Model Behavior On Training Set")
```

```python
for col_name in cols:
    f = plt.figure(figsize=(12,6))
    plt.scatter(X_test[col_name], y_test, color = "blue")
    plt.scatter(X_test[col_name], model.predict(X_test), color = "black")
    
    plt.legend(['True testing','Predicted testing'])
    plt.xlabel(col_name)
    plt.ylabel('AVG_MATH_8_SCORE')
    plt.title("Model Behavior on Testing Set")
```

<h2> Summary </h2>

The intercept of the model is approximately 259, and the coefficients for the predictor variables are as follows: 1.12 for "EXPENDITURE_PER_ENROLL", 12.82 for "INSTRUCTION_EXPENDITURE_PROPORTION",  12.47 for "FEDERAL_REVENUE_PROPORTION", and 1.28 for "PERCENT_CHANGE_MATH_8_SCORE". The R-squared score, which measures the goodness of fit, is 0.498. I find this better than expected, but can be improved based on the predictors chosen.

Examining the mean prediction error, we find that the model has an average error of approximatel1.4537, which suggests that, on average, the predictions deviate by this amount from the actual values. Additionally, the root mean squared error (RMSE) is approximatel7.5096, indicating the average magnitude of the prediction error These values indicate that the prediction error is relatively high, and that the predictions are not very close to the actual scores.

To visually assess the model's behavior, scatter plots were generated for each predictor variable against the "AVG_MATH_8_SCORE". There appears to be a slight positive correlation between "EXPENDITURE_PER_ENROLL" and "AVG_MATH_8_SCORE", implying that as expenditure per enrollment increases, math scores tend to go up slightly. Conversely, there seems to be a slight negative correlation between "FEDERAL_REVENUE_PROPORTION" and "AVG_MATH_8_SCORE", suggesting that a higher proportion of federal revenue (implying a lower proportion of local and state funding) may be associated with slightly lower math scores. 

It is difficult to find a correlation between "INSTRUCTION_EXPENDITURE_PROPORTION" and "AVG_MATH_8_SCORE", showing that there may not be a correlation between the proportion of expenditure on teachers and instructors and math scores. Finally, it seems there is a clear positive correlation between "PERCENT_CHANGE_MATH_8_SCORE" and "AVG_MATH_8_SCORE", which indicates that years in which students show positive year-over-year improvement in math scores are generally associated with higher average math scores for that year.
